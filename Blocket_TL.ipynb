```python
# Get all filters from blocket's API 
response = requests.get("https://api.blocket.se/motor-search-service/v4/search/car/filters?%2Fstart=")
```

```python
data = response.json()
```

```python
data['attributes'][3]['values']
```

```python
# Extract makes and models into a list of dictionaries
makes_and_models = []
for make in data['attributes'][3]['values']:
    if isinstance(make, dict) and 'subGroup' in make:
        for model in make['subGroup']:
            makes_and_models.append({
                'make': make['value'],
                'model': model['value'],
                'count': model['count'],
            })
```

```python
makes_and_models
```

```python
# Get total count from the quantities listed within each model
total_count = 0
for make in data['attributes'][3]['values']:
    if isinstance(make, dict) and 'subGroup' in make:
        for model in make['subGroup']:
            total_count += model['count']

print(total_count)
```

```python
# Loop through each unique make and add it's respective models 
makes_with_models = []

for make in set([entry['make'] for entry in makes_and_models]):
    models = [entry['model'] for entry in makes_and_models if entry['make'] == make]
    makes_with_models.append({'make': make, 'models': models})
```

```python
# Save the makes_with_models list as a JSON file
with open('makes_with_models.json', 'w') as file:
    json.dump(makes_with_models, file, indent=4)
```

```python
makes_with_models = []

# Load the makes_with_models list from a JSON file
with open('makes_with_models.json', 'r') as file:
    makes_with_models = json.load(file)
```

```python
# If makes_with_models is already clean
cleaned_makes_with_models = makes_with_models
```

```python
# Clean the makes_with_models list to remove models that are the same as the make
cleaned_makes_with_models = [
    {'make': entry['make'], 'models': [model for model in entry['models'] if model != entry['make']]}
    for entry in makes_with_models
]
```

```python
cleaned_makes_with_models
```

```python
# Function to extract make and model
def extract_make_and_model(heading):
    
    # Normalize the heading to NFC form (composed form)
    heading = unicodedata.normalize('NFC', heading)
    
    # First loop: Try to find both make and model
    for entry in cleaned_makes_with_models:
        make = entry['make'].lower()
        if re.search(r'\b' + re.escape(make) + r'\b', heading.lower()):
            for model in entry['models']:
                model_normalized = unicodedata.normalize('NFC', model.lower().replace('-', '').replace('.', ''))  # Normalize model and remove hyphens and periods
                if re.search(r'\b' + re.escape(model_normalized) + r'\b', heading.lower().replace('-', '').replace('.', '')):
                    return make, model_normalized.lower()
    # Second loop: If no model is found, look for just the make
    for entry in cleaned_makes_with_models:
        make = entry['make'].lower()
        if re.search(r'\b' + re.escape(make) + r'\b', heading.lower()):
            return make, None
    # If no make is found
    return None, None
```

```python
# Apply the function to create new columns
df['make'], df['model'] = zip(*df['heading'].apply(extract_make_and_model))
```
